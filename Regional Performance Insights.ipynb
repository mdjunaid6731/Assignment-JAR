<h1>Part 3: Regional Performance Insights</h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
List_of_Orders = pd.read_csv("DATASET/List_of_Orders_55FFC79CF8.csv")
Order_Details = pd.read_csv("DATASET/Order_Details_19795F61CF.csv")
```

```python
merged_df = pd.merge(List_of_Orders, Order_Details, on='Order ID', how='inner')
merged_df.head(10)
```

<h3>(a) From the List of Orders dataset, identify the top 5 states with the highest order
count. For each of these states, calculate the total sales and average profit.</h3>

```python
# top 5 states with highest order count
top_states = merged_df.groupby("State")["Order ID"].nunique().nlargest(5)
```

```python
top_states
```

```python
plt.figure(figsize=(10, 5))
top_states.plot(kind='bar', color='blue')

plt.xlabel("State")
plt.ylabel("Unique Order Count")
plt.title("Top 5 States with Highest Order Count")
plt.xticks(rotation=45)

# Add value labels on top of bars
for i, count in enumerate(top_states):
    plt.text(i, count + 1, str(count), ha='center', fontsize=12, fontweight='bold')

# Show plot
plt.show()
```

```python
top_states_list = top_states.index  # Extracting top state names
filtered_df = merged_df[merged_df["State"].isin(top_states_list)]  # Filter data
```

```python
statewise_summary = filtered_df.groupby("State").agg(
    Unique_Orders=("Order ID", "nunique"),  # Count of unique orders
    Total_Sales=("Amount", "sum"),         # Sum of sales
    Avg_Profit=("Profit", "mean")          # Average profit per order
).reset_index()
```

```python
statewise_summary
```
